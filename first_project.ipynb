```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="whitegrid")
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn import preprocessing
```

```
data = pd.read_csv('/content/sample_data/weatherHistory.csv')
```

```
data.dropna()
```

```
num = []
```

```
data.columns = data.columns.str.strip()
data.drop('Formatted Date', axis=1, inplace=True)
```

```
for col in data.columns:
  if data[col].dtype == 'object' :
    num.append(col)
```

```
for col in num:
  print(f'{col} : {len(data[col].unique())}')
```

```
le = preprocessing.LabelEncoder()
```

```
for col in num:
  data[col] = le.fit_transform(data[col])
  print('classed found : ' , list(le.classes_))
```

```
data.duplicated().sum()
```

```
data.drop_duplicates(inplace=True)
```

```
y=data['Temperature (C)']
```

```
X=data.drop(['Temperature (C)','Apparent Temperature (C)'],axis=1)
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=44, shuffle =True)

#Splitted Data
#print('X_train shape is ' , X_train.shape)
#print('X_test shape is ' , X_test.shape)
#print('y_train shape is ' , y_train.shape)
#print('y_test shape is ' , y_test.shape)
```

```
LassoRegressionModel = Lasso(alpha=1.0,random_state=33)
LassoRegressionModel.fit(X_train, y_train)

#Calculating Prediction
y_pred0 = LassoRegressionModel.predict(X_test)
```

```
RidgeRegressionModel = Ridge(alpha=1.0,random_state=33)
RidgeRegressionModel.fit(X_train, y_train)

#Calculating Prediction
y_pred = RidgeRegressionModel.predict(X_test)
```

```
#Calculating Details
print('Lasso Regression Train Score is : ',RidgeRegressionModel.score(X_train, y_train) , LassoRegressionModel.score(X_train, y_train))
print('Lasso Regression Test Score is : ',RidgeRegressionModel.score(X_test, y_test) , LassoRegressionModel.score(X_test, y_test))
print('Lasso Regression Coef is : ',RidgeRegressionModel.coef_,'\n' , LassoRegressionModel.coef_)
print('Lasso Regression intercept is : ',RidgeRegressionModel.intercept_ , LassoRegressionModel.intercept_)
print('----------------------------------------------------')
```
